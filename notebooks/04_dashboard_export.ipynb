```python
import pandas as pd

df = pd.read_csv("../data/processed/airbnb_cleaned.csv")

print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

df.head()
```

```python
dashboard_cols = [
    "id",
    "name",
    "host_id",
    "host_name",
    "neighbourhood_group",
    "neighbourhood",
    "latitude",
    "longitude",
    "room_type",
    "price",
    "minimum_nights",
    "number_of_reviews",
    "reviews_per_month",
    "calculated_host_listings_count",
    "availability_365",
    "price_per_review",
    "is_private_room",
    "is_entire_home",
    "has_reviews"
]

dashboard_df = df[dashboard_cols].copy()

dashboard_df.head()
```

```python
def price_category(price):
    if price < 75:
        return "Budget"
    elif price < 150:
        return "Mid-range"
    elif price < 300:
        return "Premium"
    else:
        return "Luxury"

dashboard_df["price_category"] = dashboard_df["price"].apply(price_category)

dashboard_df[["price", "price_category"]].head()
```

```python
def availability_category(days):
    if days == 0:
        return "Not Available"
    elif days < 90:
        return "Low Availability"
    elif days < 180:
        return "Medium Availability"
    else:
        return "High Availability"

dashboard_df["availability_category"] = dashboard_df["availability_365"].apply(availability_category)

dashboard_df[["availability_365", "availability_category"]].head()
```

```python
def review_category(reviews):
    if reviews == 0:
        return "No Reviews"
    elif reviews < 10:
        return "Low Reviews"
    elif reviews < 50:
        return "Medium Reviews"
    else:
        return "High Reviews"

dashboard_df["review_category"] = dashboard_df["number_of_reviews"].apply(review_category)

dashboard_df[["number_of_reviews", "review_category"]].head()
```

```python
neighborhood_summary = (
    dashboard_df
    .groupby(["neighbourhood_group", "neighbourhood"])
    .agg(
        total_listings=("id", "count"),
        average_price=("price", "mean"),
        median_price=("price", "median"),
        average_availability=("availability_365", "mean"),
        total_reviews=("number_of_reviews", "sum")
    )
    .reset_index()
)

neighborhood_summary.head()
```

```python
dashboard_df.to_csv("../data/processed/tableau_airbnb_dashboard_data.csv", index=False)

neighborhood_summary.to_csv("../data/processed/tableau_neighborhood_summary.csv", index=False)

print("Tableau dashboard data saved successfully.")
print("Main dashboard shape:", dashboard_df.shape)
print("Neighborhood summary shape:", neighborhood_summary.shape)
```
